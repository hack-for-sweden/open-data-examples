```python
import urllib.request, json
urlPolis="https://polisen.se/api/events?locationname=Stockholm"
urlPolis="https://polisen.se/api/events"
with urllib.request.urlopen(urlPolis) as webURL:
     data = json.loads(webURL.read())
    
```

```python
import pandas as pd
```
